```python
##splitting dataset into train and test split
training_size=int(len(df1)*0.65)
test_size=len(df1)-training_size
train_data,test_data=df1[0:training_size],df1[training_size:len(df1)]
```

```python
print(train_data.shape)
print(test_data.shape)
train_data
```

```python
from sklearn.preprocessing import RobustScaler

# Initialize RobustScaler
scaler = RobustScaler()

# Fit and transform the training data
train_scaled = scaler.fit_transform(train_data.values.reshape(-1, 1))

# Transform the test data using the same scaler
test_scaled = scaler.transform(test_data.values.reshape(-1, 1))
```

```python
# import numpy as np

# # Assume train_data and test_data are numpy arrays
min_train_value = train_data.min()
max_train_value = train_data.max()
min_scaled_train_value = train_scaled.min()
max_scaled_train_value = train_scaled.max()
print(min_train_value)
print(max_train_value)
print(min_scaled_train_value)
print(max_scaled_train_value)
min_scaled_test_value = test_scaled.min()
max_scaled_test_value = test_scaled.max()
print(min_scaled_test_value)
print(max_scaled_test_value)
```

```python
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_scaled, time_step)
X_test, ytest = create_dataset(test_scaled, time_step)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
### Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=100,batch_size=64,verbose=1)
```

```python
import tensorflow as tf
```

```python
### Lets Do the prediction and check performance metrics
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```
